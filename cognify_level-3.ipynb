**Task3**

**Price Range vs. Online Delivery and Table Booking**

- Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

- Determine if higher-priced restaurants are more likely to offer these services.


```python
df2=Data_set[["Price range","Has Online delivery","Has Table booking"]]
df2
```

```python
colm=df2.columns
colm[1:]
```

```python
df2=Data_set[["Price range","Has Online delivery","Has Table booking"]]
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in colm[1:]:
    le.fit_transform(df2[i])
df2
```

```python
indx=df2["Price range"]
col=[df2["Has Online delivery"],df2["Has Table booking"]]
relation_df=pd.crosstab(indx,col)
relation_df
```

```python
relation_df.plot(kind="bar")
plt.show()
```
