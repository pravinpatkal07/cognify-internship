**Analyze the ratings and popularity of different restaurant chains.**

```python
restaurent_list=[]
avg_rate_list=[]
votes_list=[]
for i in Data_set['Restaurant Name'].unique():
    condition=Data_set['Restaurant Name']==i
    avg_rating_sum=Data_set['Aggregate rating'][condition].sum()/len(Data_set[condition])
    votes_list.append(Data_set['Votes'][condition].sum())
    restaurent_list.append(i)
    avg_rate_list.append(avg_rating_sum)
avg_ratings_pop_df=pd.DataFrame(zip(restaurent_list,avg_rate_list,votes_list),columns=['Restaurant name','Average rating','Votes'])
avg_ratings_pop_df
```
